### Q9 Convert 'Order Date' and 'Ship Date' Columns to DateTime Format.

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],errors="coerce")
df["Ship Date"]=pd.to_datetime(df["Ship Date"],errors="coerce")
```

# Level 2: Data Cleaning & Feature Engineering

### Q10. Fill Missing Values with Suitable Values.

```python
df["Order Date"] = df["Order Date"].ffill()
```

```python
df["Order Date"].isnull().sum()
```

```python
df["Ship Date"]=df["Ship Date"].ffill()
```

```python
df["Ship Date"].isnull().sum()
```

```python
df["Postal Code"]=df["Postal Code"].fillna(df["Postal Code"].mode()[0])
```

```python
df["Postal Code"].isnull().sum()
```

```python
df["Customer Name"]=df["Customer Name"].str.title()
```

```python
df.head()
```

### Q11. Remove Extra Spaces from Product Names.

```python
df["Product Name"]=df["Product Name"].str.strip()
```

### Q12. Create a new column 'Profit Status' (Profit / Loss).

```python
df["Profit Status"]=df["Profit"].apply(lambda x:"Profit" if x>0 else("Loss" if x<0 else "No Profit Loss"))
```

### Q13. Round Off the Sales Values to Two Decimal Places.

```python
df["Sales"]=df["Sales"].round(2)
```

### Q14. Create a new column named 'Discount Percentage'.

```python
df["Discount Percentage"] = (df["Discount"] * 100).astype(int)
```

```python
print(df["Discount Percentage"])
```

###  Q15. Extract Year, Month  Order Date.

```python
df["Order Year"]=df["Order Date"].dt.year
df["Order Month"]=df["Order Date"].dt.month_name()
print(df["Order Year"])
```

# Level 5: Pivot Table Analysis

### Q33. Create a Pivot Table to display Category-wise Total Sales across Regions.

```python
Category_Region_Sales = pd.pivot_table(
    df,
    index="Category",
    columns="Region",
    values="Sales",
    aggfunc="sum",
    fill_value=0
)

print(Category_Region_Sales)
```

### Q34. Create a Pivot Table to display Category-wise Total Sales across Segments.

```python
Category_Seg_Sales = pd.pivot_table(
    df,
    index="Category",
    columns="Segment",
    values="Sales",
    aggfunc="sum",
    fill_value=0

)

print(Category_Seg_Sales )
```

### Q35. Create a Pivot Table to display Sub-Category-wise Total Sales across Regions.

```python
SubCategory_Sales = pd.pivot_table(
    df,
    index="Sub-Category",
    columns="Region",
    values="Sales",
    aggfunc="sum",
    fill_value=0

)

print(SubCategory_Sales )
```

### Q36.  Create a Pivot Table to display Region-wise Average Discount.

```python
region_Avg_Dis = pd.pivot_table(
    df,
    index="Region",
    values="Discount",
    aggfunc="mean",
    fill_value=0
)

print(region_Avg_Dis)
```

### Q37.  Create a Pivot Table to display Ship Mode-wise Order Count across Categories.

```python
ship_mode_category = pd.pivot_table(
    df,
    index="Ship Mode",
    columns="Category",
    values="Order ID",
    aggfunc="count",
    fill_value=0
)

print(ship_mode_category)
```

# Business Insights

### Q41. Find the Top 3 Customers from each Region based on Total Sales.

```python
top3_customers = (
    df.groupby(["Region", "Customer Name"], as_index=False)["Sales"]
      .sum()
)

top3_customers = (
    top3_customers
    .sort_values(["Region", "Sales"], ascending=[True, False])
    .groupby("Region")
    .head(3)
)

print(top3_customers)
```

### Q42. Calculate the Running Total of Sales.

```python
df = df.sort_values("Order Date")

df["Running Total Sales"] = df["Sales"].cumsum()

print(df[["Order Date", "Sales", "Running Total Sales"]])
```

### Q43. Rank Sub Category based on Total Sales.

```python

Subcategory_sales = (
    df.groupby("Sub-Category",as_index=False )["Sales"]
      .sum()
)

Subcategory_sales["Rank"]=Subcategory_sales["Sales"].rank(
    method="dense",
    ascending=False)
Subcategory_sales=Subcategory_sales.sort_values(by="Rank",ignore_index=True)
print(Subcategory_sales)
```

### Q44. Calculate the Percentage Contribution of Each Segment to Total Profit.

```python
segment_profit=df.groupby("Segment",as_index=False)["Profit"].sum()
total_profit=segment_profit["Profit"].sum()
segment_profit["Contribution (%)"]=(segment_profit["Profit"]/total_profit)*100
print(segment_profit)
```

### Q45. Find the Most Frequently Ordered Product.

### Q46.Find the Customer Retention Rate (if multiple years are available).

```python
customer_year = (
    df.groupby("Customer Name")["Order Year"]
      .nunique()
      .reset_index()
)

retained_customers = customer_year[customer_year["Order Year"] > 1]

print(retained_customers)
```

### Q47. Calculate the Profit Margin (%) for each Order.

```python
df["Profit Margin"]=(df["Profit"]/df["Sales"])*100
df["Profit Margin"].round(2)
```
